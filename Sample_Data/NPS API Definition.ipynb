```python
import requests
import pandas as pd 

def NPS_API_call(api_key, url):
    params = {'api_key': api_key}
    r = requests.get(url, params=params)
    print(r)
    txt = r.json()
    print(txt)
    df = pd.DataFrame(txt['data']) #example - ammend to fit returned data structure
    return df
```

```python
df = NPS_API_call(api_key='My_API', url='https://developer.nps.gov/api/v1/alerts?')
df.to_csv("Alerts.csv", index=False)
df
```

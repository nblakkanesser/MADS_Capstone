```python
import pandas as pd 
import json
import os
import openai
from openai import OpenAI
import time
import sys

sys.path.insert(0,'../')
from environment import env
from environment import helper_functions
```

#### Define Variables

```python
config = env.env()

# Inputs
client = OpenAI(api_key  = config['gpt_api_key'])
gpt_parkcode_model = config['gpt_parkcode_model']
gpt_endpoint_model = config['gpt_endpoint_model']
test_data = 'parkcode_test_data.jsonl'

# Output
```

#### Read Inputs

```python
parkcode_model = client.fine_tuning.jobs.retrieve(gpt_parkcode_model).fine_tuned_model
endpoint_model = client.fine_tuning.jobs.retrieve(gpt_endpoint_model).fine_tuned_model
```

#### Test Fine Tuning

```python
with open(test_data, 'r') as f:
    for line in f:
        data_dict = json.loads(line.strip())
        query= data_dict['prompt']
        max_tokens = 3
        endpoint = helper_functions.handle_query(query,endpoint_model,client,max_tokens)
        max_tokens = 5
        parkcode = helper_functions.handle_query(query,parkcode_model,client,max_tokens)
        print(f"query: {query}\n {endpoint} \n {parkcode}")
```
